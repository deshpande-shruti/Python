### Filter

```python
nums = [3, 2, 6, 7, 8, 4, 6, 2, 9]
```

```python
def is_even(n):
    return n % 2 == 0
```

```python
evens = list(filter(is_even, nums))
```

```python
evens
```

### Decorators

Changing the function without even touching it

```python
def div(a, b):
    print(a/ b)
```

```python
div(2, 4)
```

```python
def div(a, b):
    if a < b:
        a, b = b, a
    print(a/ b)
```

```python
div(2, 4)
```

```python
def div(a, b):
    print(a/ b)
    

# Decorator
def smart_div(func):
    
    def inner(a, b):
        if a < b:
            a, b = b, a
        return func(a, b)
    
    return inner
```

```python
div = smart_div(div)
```

```python
div(2, 4)
```

### args and kwargs

```python
def person(name, *data):
    print(name)
    print(data)
```

```python
p1 = person("Aditya", 16, "Bangalore", 9268268)
```

```python
def person(name, *data):
    print(name)
    for i in data:
        print(i)
```
